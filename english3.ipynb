```python
import pandas as pd
import spacy
import pyinflect
import random   
```

```python
text2 =  '''
Little Red Cap

Jacob and Wilhelm Grimm

Once be know upon a time went there was a sweet little girl. Everyone who saw her liked her, but most of all her grandmother, who did not know what to give the child next. Once she gave her a little cap made of red velvet. Because it suited her so well, and she wanted to wear it all the time, she came to be known as Little Red Cap.
One day her mother said to her, "Come Little Red Cap. Here is a piece of cake and a bottle of wine. Take them to your grandmother. She is sick and weak, and they will do her well. Mind your manners and give her my greetings. Behave yourself on the way, and do not leave the path, or you might fall down and break the glass, and then there will be nothing for your sick grandmother."

Little Red Cap promised to obey her mother. The grandmother lived out in the woods, a half hour from the village. When Little Red Cap entered the woods a wolf came up to her. She did not know what a wicked animal he was, and was not afraid of him.
    '''
```

```python
import nltk
from nltk.corpus import stopwords
tokens_sens = nltk.tokenize.sent_tokenize(text2, language='english')
#Создаем датафрейм
df_sentences = pd.DataFrame({'sentence': tokens_sens})
#st.write(df_sentences)

from nltk.tokenize import sent_tokenize
  
text = "Hello everyone. Welcome to GeeksforGeeks. You are studying NLP article"
sent_tokenize(text)

exercise_type = 'Выберите правильную форму глагола'
#exercise_type = 'Выбор правильного прилагательного'

#exercise_type =  'Выберите правильный артикль'

options = []
task = ''
answer = ''

df = pd.DataFrame({'sentence':'', 'options': options, 'answer':answer, 'result':[], 'task':task})
nlp = spacy.load("en_core_web_sm") # изменение формы глагола с помощью pyinflect
for sentence in df_sentences.sentence:
    #print(len(sentence))
        #st.write(sentence)
    # options = ['a', 'the', 'an']
    # answer = [i for i in [' a ', ' the ', ' an '] if i in sentence]
    



    
    #if any(word in sentence for word in [' a ', ' the ', ' an ']):
    #    print(word)
    #for i in sentence:
    #    print(i)
    #matching = [x for ' a ' in sentence if x in sentence]

    for token in nlp(str(sentence)):
        
        # if token.text == 'the':
        #     print(sentence)
        
        # if str(sentence) in  ['a', 'the', 'an'] and exercise_type == 'Выберите правильный артикль':
        #     options = ['a', 'the', 'an']
        #     answer = token
        #options = []
        if token.pos_=='VERB' and exercise_type == 'Выберите правильную форму глагола':
            answer = [token.text for token in nlp(str(sentence)) if token.pos_=='VERB']
            #options = [token._.inflect('VB'), token._.inflect('VBN'), token._.inflect('VBP'), token._.inflect('VBZ'), token._.inflect('VBG'), token._.inflect('VBD')]
            #s1, s2 =[token._.inflect('VB'), token._.inflect('VBN') for token in nlp(str(sentence)) if token.pos_=='VERB']
            # for i in answer:
            #     print(i)
            #     print(sent_tokenize(i)._.inflect('VB'))
            #print(list(set(['fsfd', 'dsfsdf' ,1 ,3 ,4, 1])))
            #print(set([token._.inflect('VB'), token._.inflect('VBN'), token._.inflect('VBP'), token._.inflect('VBZ'), token._.inflect('VBG'), token._.inflect('VBD')]))
            options.append(list(set([token._.inflect('VB'), token._.inflect('VBN'), token._.inflect('VBP'), token._.inflect('VBZ'), token._.inflect('VBG'), token._.inflect('VBD')])))
            task = token.pos_
            #print(options)
 #           {"be": {"VB": ["be"], "VBP": ["am", "are"], "VBN": ["been"], "VBD": ["was", "were"], "VBG": ["being"], "VBZ": ["is"]}}
#{"beach": {"VBN": ["beached"], "VB": ["beach"], "VBG": ["beaching"], "NN": ["beach"], "VBP": ["beach"], "VBZ": ["beaches"], "VBD": ["beached"], "NNS": ["beaches"]}}
            
            
        elif token.pos_=='ADJ'and exercise_type == 'Выбор правильного прилагательного':
            options = [token, token._.inflect('JJS'), token._.inflect('JJR'), token._.inflect('JJ'), token._.inflect('RB'), token._.inflect('RBR'), token._.inflect('RBS')]
            answer = token
            task = token.pos_ 
        else: pass
# отобрать предложения, где есть глагол или прилагательное
# отфильтровать короткие            
    # print(str(token))
    
    # sentence = sentence.replace(str(token), '___')
    # print(sentence)            
    if len(nlp(str(sentence))) > 3:
        df.loc[len(df)]=[sentence, options, answer, [], task]  
        options=[]  
        #print(sentence)
    #df['sentence'] = df['sentence'].apply(lambda x: x.replace(str(answer), ' ___ '))
    
```
